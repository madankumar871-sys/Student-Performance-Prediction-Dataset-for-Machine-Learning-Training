```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import warnings
#Suppress warnings
warnings.filterwarnings('ignore')
```

```python
#Load the wine quality dataset
url="student_dataset_10000_rows.csv"
student_data=pd.read_csv(url)

#Display the first few rows of the dataset
print("first 5 rows of the wine quality dataset:")
student_data.head()
```

```python
#Basic dataset information
print("\nDataset Information")
student_data.info()
```

```python
#Check for missing values
print("\nMissing values per columns:")
student_data.isnull().sum()
```

```python
#Basic Statistics
student_data.describe()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Target 1: Numerical score (Regression)
# Target 2: Categorical status (Classification)
X = student_data.drop(columns=["exam_score", "placement_status"])
y_regression = student_data["exam_score"]
y_classification = LabelEncoder().fit_transform(student_data["placement_status"])

# Split data into training (80%) and testing (20%) sets
X_train, X_test, y_train_reg, y_test_reg = train_test_split(
    X, y_regression, test_size=0.2, random_state=42
)
_, _, y_train_clf, y_test_clf = train_test_split(
    X, y_classification, test_size=0.2, random_state=42
)

# Standardize numerical features for stable model training
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

#2. Feature Engineering and Data Splitting

```python
from sklearn.metrics import accuracy_score, mean_absolute_error, r2_score
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor

# --- Task A: Predict Exam Score (Regression) ---
reg_model = RandomForestRegressor(random_state=42)
reg_model.fit(X_train_scaled, y_train_reg)
reg_preds = reg_model.predict(X_test_scaled)

print("--- Regression Metrics ---")
print("Mean Absolute Error (MAE):", mean_absolute_error(y_test_reg, reg_preds))
print("R-squared (R2) Score:", r2_score(y_test_reg, reg_preds))

# --- Task B: Predict Placement Status (Classification) ---
clf_model = RandomForestClassifier(random_state=42)
clf_model.fit(X_train_scaled, y_train_clf)
clf_preds = clf_model.predict(X_test_scaled)

print("\n--- Classification Metrics ---")
print("Accuracy:", accuracy_score(y_test_clf, clf_preds))
```

```python
import joblib

# Assuming reg_model and scaler are already trained from the previous step
joblib.dump(reg_model, "student_model.pkl")
joblib.dump(scaler, "scaler.pkl")
```
